```python
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(0,4*np.pi , 200)
N = np.linspace(1,1000 , 1000)
def f1(t,N):
    print("N = " + str(N))
    Sum = np.zeros(200)
    for i in range(1,N+1 , 2):
        Sum += np.sin(i*T)/i
        
    return Sum

def f2(t,N):
    print("N = " + str(N))
    Sum = np.zeros(200)
    for i in range(1,N+1):
        Sum += np.sin(i*T)/i
        
    return Sum
```

```python
plt.plot(T , f1(T,1))
plt.xlabel("f1(t)")
plt.show()
plt.plot(T , f1(T,5)) 
plt.xlabel("f1(t)")
plt.show()
plt.plot(T , f1(T,2000))
plt.xlabel("f1(t)")
plt.show()
```

Plots of the even function obtained by modifiying the given summation as shown above

```python
def f1(t,N , A=1, D=0.5):
    print("N = " + str(N))
    Sum = np.zeros(200)
    for i in range(1,N+1 , 2):
        Sum += A*(np.sin(i*T) - np.sin(i*(T-2*np.pi*D))) /i
        
    return Sum
def f2(t,N , A=1 , D=0.5):
    print("N = " + str(N))
    Sum = np.zeros(200)
    for i in range(1,N+1):
        Sum += A*(np.sin(i*T) - np.sin(i*(T-2*np.pi*D)))/i
        
    return Sum
```

Modified f1(t) and f2(t) to adjust amplitude to A and duty cycle to D. The same is plotted below with Duty Cycle = 25% and Amplitude = 2

```python
A = 2
D = .25 
plt.plot(T , f1(T,1 ,A,D))
plt.xlabel("f1(t)")
plt.show()
plt.plot(T , f1(T,5 , A , D)) 
plt.xlabel("f1(t)")
plt.show()
plt.plot(T , f1(T,2000 , A ,D))  
plt.xlabel("f1(t)")
plt.show()
```
