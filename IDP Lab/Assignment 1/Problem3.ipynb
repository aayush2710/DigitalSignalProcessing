```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 401
t = np.linspace(-2 , 2 , N)
w = np.linspace(-np.pi , np.pi , N)

def f(t,a=1):
    return np.exp(-(a**2)*np.pi*t**2)
def g(t):
    return np.cos(2*np.pi*t) + 0.5*np.sin(4*np.pi*t)

def Ff(s,a=1):
    print("a = " + str(a))
    return np.exp(-(a**-2)*np.pi*s**2)/a**2

plt.plot(t,np.abs(Ff(t,1)))
plt.xlabel('s')
plt.ylabel('|Ff(s)|')
plt.show()
plt.plot(t,np.abs(Ff(t,100)))
plt.xlabel('s')
plt.ylabel('|Ff(s)|')
plt.show()
```

```python
Rtime = 0.1
def p(t , Rtime):
    L=[]
    for i in t:
       
        if abs(i/Rtime - round(i/Rtime)) < 0.01:
            L.append(1)
        else:
            L.append(0)
    return np.array(L)
    

plt.plot(t,p(t , Rtime))
plt.show()
print("Impulse Train P(t) for sampling")
            
```

```python
x = f(t)*p(t,Rtime)
q = int(2/Rtime)
x1 = []
N = np.linspace(-q,q , (2*q)+1)
for i in x:
    
    if i!=0:
        x1.append(i)
x = np.array(x1)

plt.scatter(N , x)
plt.xlabel('n')
plt.ylabel('f[n]')
plt.show()
print("Sampled Signal")
```

```python
L = 3.2
n = L/(2*Rtime)

#print(p)
n = int(n)
#print(x)
x1 = x[q+1-n:q+1+n]
N1 = N[q+1-n:q+1+n]
plt.scatter(N1,x1)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()
print("Signal after applying the window function")
```

```python
B = 3.2
Rfreq = B*Rtime/L

m = B/Rfreq + 1
#print(m)
fx = np.fft.fftn(x1,int(m))
#print(fx)
m = int(m)
plt.scatter((2*np.pi/m)*np.linspace(-int(m//2) , int(m//2) , int(m)) , np.abs(fx))
plt.xlabel('w')
plt.ylabel('Ff[w]')
plt.show()
print("FT of sampled signal further sampled in frequency domain")
```

```python

def w2(t):
    k = []
    for i in t:
        
    
        
        if abs(i)<= L/2:
            
            k.append(1- (2*abs(i)/L))
        else:
            k.append(0)
            
    return np.array(k)

def w3(t):
    k = []
    for i in np.sin(2*np.pi*t/L)**2:
        
    
        
        if abs(i)<= L/2:
            
            k.append(i)
        else:
            k.append(0)
            
    return np.array(k)

f2 = f(t)*w2(t)
f3 = f(t)*w3(t)
  
  
    
```

```python
y2 = f2*p(t,Rtime)
y3 = f3*p(t,Rtime)
x1 = []
N = np.linspace(-10,10 , 21)
for i in y2:
    
    if i!=0:
        x1.append(i)
y2 = np.array(x1)
x1 = []
N = np.linspace(-10,10 , 21)
for i in y3:
    
    if i!=0:
        x1.append(i)
y3 = np.array(x1)
```

```python
plt.scatter((2*np.pi/m)*np.linspace(-m/2 , m/2 , m) , np.abs(np.fft.fft(y2,int(m))))
plt.xlabel('w')
plt.ylabel('Ff[w]')

plt.show()
print("DFT with window function w2")
plt.scatter((2*np.pi/m)*np.linspace(-m/2 , m/2 , m) , np.abs(np.fft.fft(y3,int(m))))
plt.xlabel('w')
plt.ylabel('Ff[w]')

plt.show()
print("DFT with window function w3")
```
