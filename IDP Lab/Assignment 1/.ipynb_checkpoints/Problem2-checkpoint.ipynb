```python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(-4,4,200)
from math import floor
def f1(t):
    L = []
    for j in t:
        i = floor(j-0.5)
        if i==floor(j-1):
            L.append(1)
        elif i==floor(j):
            L.append(-1)
        else:
            L.append(0)
            
    return np.array(L)

def f2(t):
    L = []
    for j in t:
        i = floor(j-0.5)
        if i==floor(j-1):
            L.append(0.5 - j + round(j))
        elif i==floor(j):
            L.append(0.5+j - round(j))
        else:
            L.append(0)
            
    return np.array(L)


def fft(x):
    return np.fft.fft(x)   

plt.rcParams['figure.figsize'] = (15,7)
```

```python
plt.plot(A , f1(A))
plt.xlabel("f1(t)")
plt.show()

plt.plot(A , f2(A))
plt.xlabel("f2(t)")
plt.show()
```

Properties of f1(t)

1) Discontinuous

2) ODD

3) Possess half wave symmetry

    => f1(t - 0.5) = -f1(t) 
    
    
4) Possess quarter wave symmetry

    => f1 is half symmetric
    
    => f1 is even symmetric about 0.25(i.e 1/4th of the period)
    

Properties of f2(t)

1) Continuous

2) EVEN

3) Does not Possess half wave symmetry

=> f1(t - 0.5) != -f1(t) 

4) Does not Possess quarter wave symmetry

=> f1 is not half symmetric


```python
print("Fourier Spectrum of f1(t)")
plt.plot(np.real((fft(f1(A)) + fft(f1(-A)))/2))
plt.xlabel("k")
plt.ylabel("ak")
plt.show()

plt.plot(np.real((fft(f2(A)) + fft(f2(-A)))/2))
print("Fourier Spectrum of f2(t)")
plt.xlabel("k")
plt.ylabel("ak")
plt.show()
```

                                    ### Plots of ak

```python
print("Fourier Spectrum of f1(t)")
plt.plot(np.real((fft(f1(A)) - fft(f1(-A)))/-2j))
plt.xlabel("k")
plt.ylabel("bk")
plt.show()

plt.plot(np.real((fft(f2(A)) - fft(f2(-A)))/-2j))
print("Fourier Spectrum of f2(t)")
plt.xlabel("k")
plt.ylabel("ak")
plt.show()
```

                                    ### Plots of bk

Calculated using

ak = (ck + c-k)/2

bk = (ck - c-k)/-2j

```python


print("Fourier Spectrum of f1(t)")
plt.plot(np.abs(fft(f1(A))))
plt.xlabel("k")
plt.ylabel("ck")
plt.show()

plt.plot(np.abs(fft(f2(A))))
print("Fourier Spectrum of f2(t)")
plt.xlabel("k")
plt.ylabel("ck")
plt.show()
```

                                    ### Plots of Ck

The fourier series converges for the functions f1(t) and f2(t).

The given functions f1(t) and f2(t) are non differentiable have corner points. Thats why they can be approximated by high frequency sinusoids only. As we did in problem 1 we tried to approximate f1 and f2 as summation of sinusoids we noticed that for higer value of N we got more and more close to the required function.
